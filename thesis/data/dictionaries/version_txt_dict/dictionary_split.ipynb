```python
affect_dict_path = 'Affect.txt'
with open(affect_dict_path, 'r') as file:
    affect_dict = file.readlines()
    # Remove newline characters
    affect_dict = [line.strip() for line in affect_dict]
```

```python
affect_dict
```

```python
moral_dict_path = 'Moral.txt'
with open(moral_dict_path, 'r') as file:
    moral_dict = file.readlines()
    # Remove newline characters
    moral_dict = [line.strip() for line in moral_dict]
```

```python
moral_dict
```

```python
# Convert the datasets into sets
moral_set = set(moral_dict)
affect_set = set(affect_dict)

# Find the common words by taking the intersection of the sets
moral_emotional_dict = list(moral_set.intersection(affect_set))
```

```python
moral_emotional_dict
```

```python
only_moral_dict = list(moral_set.difference(affect_set))
```

```python
only_moral_dict
```

```python
only_affect_dict = list(affect_set.difference(moral_set))
```

```python
only_affect_dict
```

```python
with open("moral_emotional_dict.txt", "w") as file:
    for item in moral_emotional_dict:
        file.write(item + "\n")
```

```python
with open("only_affect_dict.txt", "w") as file:
    for item in only_affect_dict:
        file.write(item + "\n")
```

```python
with open("only_moral_dict.txt", "w") as file:
    for item in only_moral_dict:
        file.write(item + "\n")
```
